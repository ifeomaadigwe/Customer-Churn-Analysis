### What is Customer Churn?

Customer churn, also known as customer attrition, refers to the phenomenon where customers discontinue their relationship 
or cease doing business with a company. 
Customer churn is also referred to the rate at which customers stop using a company's products or services over a 
specific period. It is measured by tracking the number or percentage of customers who discontinue their engagement 
with the company.



### Some reasons for Customer Churn:

- Poor Customer Experience: Customers may churn due to unsatisfactory experiences, such as lack of responsiveness, 
product/service quality issues, or inadequate support.

- Competitive Factors: Strong competition in the market can lure customers away, offering better pricing, features, or 
customer service.

- Lack of Value: If customers do not perceive sufficient value or benefits from a product or service, they are more 
likely to churn.

- Changing Needs or Circumstances: Customers' requirements or circumstances may evolve over time, making the company's 
offerings less relevant.

- Pricing or Contractual Factors: Uncompetitive pricing, unfavorable contract terms, or unexpected fee increases can 
lead to customer dissatisfaction and churn.

# EDA - Exploratory Data Analysis

```python
# Get the total number of customers
total_customers = len(data)

# Print the total number of customers
print("Total number of customers:", total_customers)
```

```python
# Count of active customers and non-active customers
active_customers = data[data['Churn'] == 'No'].shape[0]
non_active_customers = data[data['Churn'] == 'Yes'].shape[0]

# Print the count of active and non-active customers
print("Number of active customers (Non-churned):", active_customers)
print("Number of non-active customers (Churned):", non_active_customers)
```

```python
# What is the distribution of churned vs non-churned customers?
# Count the number of churned and non-churned customers
churn_counts = data['Churn'].value_counts()

# Create a pie chart
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.2f%%', startangle=90)
plt.title('Distribution of Churned vs Non-Churned Customers')
plt.axis('equal')
plt.show()
```

```python
# What is the distribution of customer churn across different contract types?

# Calculate churn rate by contract type
churn_rate_by_contract_type = data.groupby('Contract')['Churn'].mean()

# Print the churn rate by contract type
print("Churn rate by contract type:")
print(churn_rate_by_contract_type)

# Visualize the churn rate by contract type
sns.countplot(x='Contract', hue='Churn', data=data)
plt.title('Customer Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.show()
```

###### Churn Rate by Contract Type:

- Customers with month-to-month contracts have the highest churn rate of approximately 42.7%.
- Customers with one-year contracts have a relatively lower churn rate of around 11.3%.
- Customers with two-year contracts have the lowest churn rate of only 2.8%.

```python
# How does customer churn vary based on gender?

# Calculate churn counts by gender
churn_counts = data.groupby(['gender', 'Churn']).size().unstack()

# Plot the churn rate by gender
sns.set_palette("colorblind")  # Optional: Set a color palette
ax = churn_counts.plot(kind='bar', stacked=False)

# Add value labels to each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

# Set the plot labels and title
plt.title('Customer Churn by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

# Show the plot
plt.legend(['Non-Churned', 'Churned'])
plt.show()
```

###### Churn Rate by Gender:

- Both male and female customers have similar churn rates.
- Among active customers, there are 2,549 females and 2,625 males.
- Among non-active customers (churned), there are 939 females and 930 males.

```python
# Churn rate by pament method

# Convert 'Churn' column to numeric (0 for 'No', 1 for 'Yes')
data['Churn'] = data['Churn'].map({'No': 0, 'Yes': 1})

# Calculate churn rate by payment method
churn_rate_by_payment_method = data.groupby('PaymentMethod')['Churn'].mean()

# Print the churn rate by payment method
print("Churn rate by payment method:")
print(churn_rate_by_payment_method)

# Visualize the churn rate by payment method
sns.barplot(x=churn_rate_by_payment_method.index, y=churn_rate_by_payment_method.values)
plt.title('Churn Rate by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Churn Rate')
plt.xticks(rotation=45)
plt.show()
```

- Customers who choose electronic check as their payment method exhibit the highest churn rate at 45.28%
- Customers who opted for credit card as their payment method exhibit the lowest churn rate at 15.24%

```python
# How does the choice of payment method influence customer churn?

# Visualize the churn rate by payment method
sns.countplot(x='PaymentMethod', hue='Churn', data=data)
plt.title('Customer Churn by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

###### Churn Rate by Payment Method:

- Customers using electronic checks as their payment method have the highest churn rate of approximately 45.3%.
- Customers using credit card (automatic) or bank transfer (automatic) have relatively lower churn rates of around 15.2% and 16.7% respectively.
- Customers using mailed checks have a churn rate of approximately 19.1%.

```python
# Calculate churn rate by dependents
churn_rate_by_dependents = data.groupby('Dependents')['Churn'].mean()

# Create a pie chart to visualize the dependence of churn on dependents
plt.pie(churn_rate_by_dependents, labels=churn_rate_by_dependents.index, autopct='%1.1f%%')
plt.title('Churn Rate by Dependents')
plt.show()
```

```python
# Calculate churn rate by technical support

churn_rate_by_tech_support = data.groupby('TechSupport')['Churn'].mean()

# Create a bar chart to compare churn rate by technical support
sns.barplot(x=churn_rate_by_tech_support.index, y=churn_rate_by_tech_support.values)
plt.title('Churn Rate by Technical Support')
plt.xlabel('Technical Support')
plt.ylabel('Churn Rate')
plt.show()
```

- The presence or absence of technical support plays a significant role in customer churn rates. Customers who have access to 
technical support (Yes) exhibit a churn rate of 15.17%, indicating a lower likelihood of churn compared to those who do not
have this support.

- On the other hand, customers who lack technical support (No) experience a significantly higher churn rate of 41.64%, suggesting 
that the absence of such support contributes to a greater likelihood of churn.

- These findings emphasize the importance of providing technical support in customer retention efforts. Customers who can rely 
on technical support are more inclined to remain loyal to the company, whereas those without access to technical support or 
internet service are at a higher risk of churn.

- Based on these observations, it would be advisable for the company to prioritize the improvement and expansion of their 
technical support services. Doing so can help reduce churn rates, enhance customer satisfaction, and foster long-term customer
relationships.


```python

# Calculate churn rate by online backup
churn_rate_by_backup = data.groupby('OnlineBackup')['Churn'].mean().reset_index()

# Sort the backup types by churn rate in descending order
churn_rate_by_backup = churn_rate_by_backup.sort_values(by='Churn', ascending=False)

# Visualize the churn rate by online backup
ax = sns.barplot(x='OnlineBackup', y='Churn', data=churn_rate_by_backup)
plt.title('Churn Rate by Online Backup')
plt.xlabel('Online Backup')
plt.ylabel('Churn Rate')
plt.xticks(rotation=45)

# Add text annotations to the bars
for index, row in churn_rate_by_backup.iterrows():
    plt.text(index, row['Churn'], '{:.2%}'.format(row['Churn']), ha='center', va='bottom')

plt.show()
```

##### Churn Rate by Online Backup:

- Customers without online backup have a higher churn rate of approximately 39.9%.
- Customers with online backup have a lower churn rate of about 7.4%.
- Customers with no internet service have a churn rate of around 21.5%.



###### Churn Rate by Online Security:

```python
# Calculate churn rate by tenure

churn_rate_by_tenure = data.groupby('tenure')['Churn'].mean()

# Create an area line chart to show churn rate by tenure
plt.fill_between(churn_rate_by_tenure.index, churn_rate_by_tenure.values, alpha=0.5)
plt.plot(churn_rate_by_tenure.index, churn_rate_by_tenure.values, marker='o')

annotation_interval = 5  # Add annotation every 5 tenure periods
for x, y in zip(churn_rate_by_tenure.index, churn_rate_by_tenure.values):
    if x % annotation_interval == 0:
        plt.text(x, y, f'{y:.2f}', ha='center', va='bottom')

plt.title('Churn Rate by Tenure')
plt.xlabel('Tenure')
plt.ylabel('Churn Rate')
plt.show()
```

- Based on the area chart above, it is observed that customers with shorter tenure are more likely to churn.

```python
# What is the distribution of monthly charges?
# What is the relationship between monthly charges and the likelihood of customer churn?

# Create a grouped box plot to compare monthly charges for churned and non-churned customers
sns.boxplot(x='Churn', y='MonthlyCharges', data=data)
plt.title('Monthly Charges Distribution for Churned vs Non-Churned Customers')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.show()
```

towards higher values, as indicated by a higher median and third quartile. However, conducting additional analysis and

# Recommendations

### Recommendations: Based on the dataset made available, the company can improve customer retention and reduce churn rates by:

#### Contract Type: 
- Encourage customers to opt for longer-term contracts, such as one-year or two-year contracts, as they have lower churn rates compared to month-to-month contracts.

#### Payment Method: 
- Provide incentives or promotions for customers to use credit card or bank transfer (automatic) payment methods, as they have lower churn rates compared to electronic checks.

#### Technical Support: 
- Enhance the availability and quality of technical support services to reduce customer churn. Customers without technical support have a significantly higher churn rate.

#### Online Backup and Security: 
- Promote the use of online backup and security services to customers, as they have lower churn rates compared to customers without these services.

#### Internet Service: 
- Focus on improving the quality and reliability of fiber optic internet service to reduce churn among customers with this service. Encourage customers without internet service to explore options and benefits of having an internet connection.

#### Customer Engagement: 
- Implement proactive customer engagement strategies, such as personalized offers, loyalty programs, and regular communication, to build stronger relationships with customers and increase customer satisfaction.
